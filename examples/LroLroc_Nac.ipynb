```python
import os
import json
import ctypes
import shapely.wkt
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

import ale
import knoten.csm as csm

from ale.drivers import base
from ale.drivers.lro_drivers import LroLrocPds3LabelNaifSpiceDriver
from ale.drivers import JsonEncoder
```

```python
# Create a CSM camera for LRO LROC NAC image

# Download the image from the PDS archives
image_dir = "data"
lrocimage = os.path.join(image_dir, "M114145549LE.IMG")
downloader = urllib.request.URLopener()
downloader.retrieve("https://pdsimage2.wr.usgs.gov/Missions/Lunar_Reconnaissance_Orbiter/LROC/EDR/LROLRC_0001/DATA/MAP/2009333/NAC/M114145549LE.IMG", lrocimage)

# Generate an ISD from the image
#lrocimage = "nacl00015d79.img"

driver = LroLrocPds3LabelNaifSpiceDriver(lrocimage)

# The SPICE kernels are only available inside this "with"
with driver as d:
    aledict = d.to_dict()

# Export the ISD (python dictionary) to a JSON ISD file
alelabel = os.path.splitext(lrocimage)[0]+".json"
json.dump(aledict, open(alelabel, "w"), cls=JsonEncoder)

camera = csm.create_csm(lrocimage)    
```

```python
# Get the footprint using the model and plot it

nlines, nsamples = aledict["image_lines"], aledict["image_samples"]

boundary = csm.generate_boundary((nlines, nsamples))

#geom = csm.generate_latlon_footprint(camera, boundary, aledict['radii']['semimajor'], aledict['radii']['semiminor'])
#wkt = geom.ExportToWkt()
#P = shapely.wkt.loads(wkt)
#P.geoms[0]  

lons, lats, alts = csm.generate_latlon_boundary(camera, boundary, aledict['radii']['semimajor'], aledict['radii']['semiminor'])

plt.rcParams["figure.figsize"] = [5, 5]
plt.axes().set_aspect('equal','datalim')
plt.plot(lons, lats)
plt.xlabel("Longitude (deg)")
plt.ylabel("Latitude (deg)")
plt.title("CSM footprint")
```
